```python
# N tane sample seç ve bu seçilen sample'ları sub sample olarak adlandır. 
# Subsample'ları decision tree ile analiz et.
# N tane decision tree ile bir tane random forest modeli oluşturuyoruz.
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv")
data.drop(["id","Unnamed: 32"],axis=1,inplace = True)
```

```python
data.diagnosis = [ 1 if each == "M" else 0 for each in data.diagnosis]
y = data.diagnosis.values
x_data = data.drop(["diagnosis"],axis=1)
```

```python
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
# %% train test split
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size = 0.15,random_state = 42)
```

```python
#%% decision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

print("decision tree score: ", dt.score(x_test,y_test))
```

```python
#%%  random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100,random_state = 1)
rf.fit(x_train,y_train)
print("random forest algo result: ",rf.score(x_test,y_test))
```

```python
# n_estimators : içinde kaç tane ağaç olacağını verir.
# score ile accuracy alıyoruz. Farklı değerlendirme metriklerine de bakacağız.
```
