```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("random_forest_regression_dataset.csv",sep = ";",header = None)
df.head()
```

```python
x = df.iloc[:,0].values.reshape(-1,1)
y = df.iloc[:,1].values.reshape(-1,1)
```

```python
# %%
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100, random_state = 42) # n_estimators : ağaç sayısı 
rf.fit(x,y)
# random state 42 demek örnekleri random seçiyoruz ancak her seferinde çalıştığında random seçersek her seferinde
# model farklı şekilde eğitilir. Biz ilk başta random seçiyoruz. Daha sonra 42 ıd gibi bişi sonraki bölmelerde de ona göre 
# aynı örnekleri bölüyor yani her defasında aynı random seçenekleri getiriyor
```

```python
print("7.8 seviyesinde fiyatın ne kadar olduğu: ",rf.predict([[7.8]]))
```

```python
x_ = np.arange(np.min(x), np.max(x), 0.01).reshape(-1, 1)
y_head = rf.predict(x_)
```

```python
# visualize
plt.scatter(x,y,color="red")
plt.plot(x_,y_head,color="green")
plt.xlabel("tribun level")
plt.ylabel("ucret")
plt.show()
```
