* bir modeli iki farklı data ile test ettiğimizde birinde 89 diğerinde 60 veriorsa mesela bu sorundur
* bu sorunu çözmek için k-fold cross validation kullanılır
* modeli eğitmek için k-fold cross validation kullanılır
* overfiting : datayı ezberlemektir. istenmedik durumdur bunu önlemek gerekir.
* emin olmak için birden fazla accurac lazım
* train ile eğitim validation ile doğrulama yapıcaz
* datamızı train ve validationa göre eğitmiş ve test etmiş oluoruz.
* K değerini train ve validationa göre ayırıyoruz
* 5'e bölsek 5 adım yapacaktık
* KNN algoritmasına normalizasyon lazımdı python kodunu yazarken ona dikkat et

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
# %% normalization
x = (x-np.min(x))/(np.max(x)-np.min(x))
```

```python
# %% train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
#%% knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)  # k = n_neighbors
```

```python
# %% K fold CV K = 10 
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = knn, X = x_train, y= y_train, cv = 10) # cross val yaparken knn algoritmasını kullan.
# cv = 10 ile train datamı 10'a böldü. Her seferinde 9'u train 1 tanesi de validation
accuracies
```

```python
print("average accuracy: ",np.mean(accuracies))
print("average std: ",np.std(accuracies))
```

```python
# yayılım az yani tutarlı  data 
# k = 3 iyi değermiş o zaman test datamızı test edebiliriz.
```

```python
#%% 
knn.fit(x_train,y_train) # fit etmeyi unutma
print("test accuracy: ",knn.score(x_test,y_test)) # K foldun k'sini genelde 10 alırlar literatürde
```

## Grid Search

![resim](33.jpg)

* hem k değerinin en uygun değerini bulucaz hem de cross validation yapıcaz
* 12 farklı parcacık ortaya cıkıyor. bunun  için grid search valıdatıon denir
* 

```python
# %% grid search cross validation for knn

from sklearn.model_selection import GridSearchCV

grid = {"n_neighbors":np.arange(1,50)}
knn= KNeighborsClassifier()

knn_cv = GridSearchCV(knn, grid, cv = 10)  # GridSearchCV  algoritma, algoritmada değişecek şeyler
knn_cv.fit(x,y)
```

```python
#%% print hyperparameter KNN algoritmasindaki K degeri
print("tuned hyperparameter K: ",knn_cv.best_params_)
print("tuned parametreye gore en iyi accuracy (best score): ",knn_cv.best_score_)
```

```python
# aşağıda lojistik regresyon için grid search yapıcaz . Farklı parametreleri nasıl tune ettiğini görücez
```

```python
# lojistik 0 ya da 1di . ırıs datasetini 3 clasttan 2 classa düşürelim
# ilk 100 değer ilk 2 classı verir 
```

```python
# %% Grid search CV with logistic regression

x = x[:100,:]
y = y[:100] 
# birden fazla grid parametresi var
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
grid = {"C":np.logspace(-3,3,7),"penalty":["l1","l2"]}  # l1 = lasso ve l2 = ridge
# c parametresi lojistik regresyon regülarizasyon parametresi olarak geçer. 
# c cok buyukse overfit aşırı ezberleme , düşükse underfit olur 
# penalty parametresi de loss functionları verir
```

```python
logreg = LogisticRegression(solver='liblinear', max_iter=1000)
logreg_cv = GridSearchCV(logreg,grid,cv = 10)
logreg_cv.fit(x,y)
```

```python
print("tuned hyperparameters: (best parameters): ",logreg_cv.best_params_)
print("accuracy: ",logreg_cv.best_score_)
# bu parameteleri kullanrak %10 datamı predict edebiliyorum
```

```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import GridSearchCV
iris = load_iris()

x = iris.data
y = iris.target

x = x[:100,:]
y = y[:100] 

x = (x-np.min(x))/(np.max(x)-np.min(x))

from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import numpy as np

grid = {
    "C": np.logspace(-3, 3, 7),
    "penalty": ["l1", "l2"]
}

logreg = LogisticRegression(solver='liblinear', max_iter=1000)
logreg_cv = GridSearchCV(logreg, grid, cv=10)
logreg_cv.fit(x_train, y_train)

print("Tuned hyperparameters (best parameters):", logreg_cv.best_params_)
print("Accuracy:", logreg_cv.best_score_)
```

```python
logreg2 = LogisticRegression(C=1, penalty="l1", solver="liblinear", max_iter=1000)
logreg2.fit(x_train, y_train)
print("score:", logreg2.score(x_test, y_test))
```
