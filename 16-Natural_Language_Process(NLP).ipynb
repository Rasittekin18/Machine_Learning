```python
# lemmatization
# maça gitmek çok güzel , maç çok iyiydi, maçı kazandık --> burdaki maça maç maçı 3'ü birden gerksiz birinin tutulması yeterli
# kelimelerin köklerini bulucaz
# lemmatazation loved => love   gitmeyecegim = > git
```

```python
import nltk as nlp
nltk.download('wordnet')
nltk.download('omw-1.4')

lemma = nlp.WordNetLemmatizer()
description = [ lemma.lemmatize(word) for word in description] # description içindeki tüm kelimeler dolan ve bunların hepsinni kökünü bul
description
```

```python
description = " ".join(description) # description içindeki tüm kelimeleri boşluk ile birleştir
description
```

* bunu 4. için yapmıştık datanın geneline yaymak lazım

```python
description_list = []
for description in data.description:
    description = re.sub("[^a-zA-Z]"," ",description)
    description = description.lower()   # buyuk harftan kucuk harfe cevirme
    description = nltk.word_tokenize(description) # cümleyi kelimelerine ayır
    #description = [ word for word in description if not word in set(stopwords.words("english"))] gereksiz kelimeleri çıkar
    lemma = nlp.WordNetLemmatizer()
    description = [ lemma.lemmatize(word) for word in description]
    description = " ".join(description)
    description_list.append(description)
    # tek cümle için öğrendiklerimizi tüm dataya uyguluyoruz 
    # her bir kelime için stopword araması yapacağından dolaı uzun sürebilir o kısmı yoruma aldık 
```

# bag of Words 
![resim](29.jpg)

* merhaba featurını yeniden oluşturmuyor
* cümleleri featurlar olacak şekilde yazmış oldum
* her satırım bir cümle, her bir sütünum da bunlarda geçen unıque kelimeler
* featurlarımdan kelimeleri olan bir çanta oluşturdum
* kelimelerin her birini featura çevirmek

```python
# %% bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer # bag of words yaratmak icin kullandigim metot
max_features = 10000 # max features 16000 cümlede 32000 kelime de olsa bu 32000 kelime içinde en çok kullanılan 5000'i seç demek. Diğer türlü algoritma sıkıntı yaşar.
count_vectorizer = CountVectorizer(max_features=max_features,stop_words = "english")
count_vectorizer
```

```python
# biraz önce 1 ve 0 lardan oluşan matrise  sparce matrix denir.
sparce_matrix = count_vectorizer.fit_transform(description_list).toarray()  # x numpy olarak oluştuğu için arraye çevirdik
# list count vectorizerda tanımladığın gibi fit et uyarla daga sonra da sparce matrixe eşitle
sparce_matrix # bunların her biri bir kelime artık 
```

```python
kelimeler = count_vectorizer.get_feature_names_out()
print(list(kelimeler))
```

```python
print("en sik kullanilan {} kelimeler: {}".format(max_features, count_vectorizer.get_feature_names_out()))
```

```python
# model kurma ve predict yapma
```

```python
y = data.iloc[:,0].values   # male or female classes  values yazıp numpy array'e çevirdik. :,0 tüm satırların 0.indeksi
x = sparce_matrix
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.1, random_state = 42)
```

```python
# %% naive bayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
#%% prediction
y_pred = nb.predict(x_test)
print("accuracy:", nb.score(x_test, y_test))

```
