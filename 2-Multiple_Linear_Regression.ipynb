* Bu datasetinde maaşı etkileyen deneyimin yanında yaş da vardır.
* Simple  Linear Regression : y = b0+b1*x  --> maas = b0 + b1*deneyim
* Multiple LR : y= b0 +b1*x+ b2*x  +...  ---> maas = b0+b1*deneyim + b2*yas
* maas : dependent variable -- Bağımlı değişken
* deneyim , yaş : independent variable ---> bağımsız değişken
* LR mantığının aynısı amac = min (MSE)  B0,B1,B2 =?
* 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("multiple_linear_regression_dataset.csv", sep= ";")
```

```python
df.head()
```

```python
x = df.iloc[:,[0,2]].values
y = df.maas.values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
multiple_linear_regression = LinearRegression()
```

```python
multiple_linear_regression.fit(x,y)
```

```python
print("b0:" , multiple_linear_regression.intercept_) 
```

```python
print("b1 , b2 :" , multiple_linear_regression.coef_) 
```

```python
#predict
multiple_linear_regression.predict(np.array([[10,35],[5,35]]))
```
